```python
# get result using pytorch liberary

import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# model from text classification in nlp from hugging face
model_name = 'cardiffnlp/twitter-roberta-base-sentiment-latest'

# for pytorch
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# fixed parameters for input data
def preprocess_text(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    return inputs

# labels mentioned in result
labels = ["cry","sad","happy"]

# final classify text function
def classify_text(text):
    inputs = preprocess_text(text)
    outputs = model(**inputs)
    logits = outputs.logits
    probabilities = torch.nn.functional.softmax(logits, dim=-1)
    predicted_class_index = torch.argmax(probabilities, dim=-1).item()
    predicted_label = labels[predicted_class_index]
    predicted_score = probabilities[0][predicted_class_index].item()
    return predicted_label, predicted_score

# input data
texts = [
    "I love this movie!",
    "This is the worst film I've ever seen.",
    "Hey, it was bad, but it should be better when the hero was on in interval, it good"
]

# loop and output
for text in texts:
    label, score = classify_text(text)
    print(f"Text: {text}")
    print(f"Label: {label}, Score: {score:.4f}")
```

```python
# get result using tensorflow library

import tensorflow as tf
import numpy as np
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification

# model from text classification from nlp in hugging face
model_name = 'cardiffnlp/twitter-roberta-base-sentiment-latest'

# basic setting for tf
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForSequenceClassification.from_pretrained(model_name)

# fixed parameters for input data
def preprocess_text(text):
  inputs = tokenizer(text, return_tensors='tf', truncation=True, padding=True, max_length=512)
  return inputs

# label names
labels = ["cry","sad","happy"]

# final function to get result
def classify_text(text):
  inputs = preprocess_text(text)
  outputs = model(**inputs)
  logits = outputs.logits
  probabilities = tf.nn.softmax(logits, axis=-1).numpy()
  predicted_class_index = np.argmax(probabilities, axis=-1)[0]
  predicted_labels = labels[predicted_class_index]
  predicted_score = probabilities[0][predicted_class_index]
  return predicted_labels, predicted_score

# input data
texts = [
    "I love this movie!",
    "This is the worst film I've ever seen.",
    "Hey, it was bad, but it should be better when the hero was on in interval, it good"
]

# loop and output
for text in texts:
    label, score = classify_text(text)
    print(f"Text: {text}")
    print(f"Label: {label}, Score: {score:.4f}")
```
